```python
import jieba
def count_word(text,worddict,date):
    wordsls=jieba.lcut(text)
    for i in wordsls:
        if i in worddict:
            worddict[i] = worddict[i]+1
    wcls=list(worddict.items())
    return date,wcls
```

```python
b = []
q = count_word(news4082.content[0],pro_dict,news4082.date[0])
b.append(q)
freq_words = pd.DataFrame(b,columns = ['date','words'])
freq_words
```

```python
list1 = list(pro_dict)
def split_list(dict,data):
    list1 = list(dict)
    data[list1] = data['words'].apply(pd.Series)
    for i in list1:
        data[[i*2,i*3]] = data[i].apply(pd.Series)
        del data[i]
        del data[i*2]
        data[i]=data[i*3]
        del data[i*3]
    return data
freq_words1 = split_list(pro_dict,freq_words)
freq_words2 = freq_words1.drop('words',axis=1)
freq_words2
```

```python
freq_words3 = np.transpose(freq_words2)
freq_words4 = freq_words3.reset_index()
c = str(freq_words2.date[0])
freq_words4.columns = ['area',c]
freq_words5 = freq_words4[1:]
freq_words5.sort_values(c,inplace=True)
freq_words6 = freq_words5[freq_words5['20210408'] > 0]
freq_words6
```

```python
import re
def get_words(text,key,date):
    pattern = re.compile(r'.*?(.{10}%s)'%key) 
    output = re.findall(pattern,text)
    return key,date,output
```
